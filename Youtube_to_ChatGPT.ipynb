```python
import sys
import traceback
from time import sleep
from urllib3.exceptions import MaxRetryError, NewConnectionError
from selenium import webdriver
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from selenium.webdriver.firefox.options import Options
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

# Set up Selenium WebDriver
options = Options()
options.set_preference("general.useragent.override", "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:96.0) Gecko/20100101 Firefox/96.0")
options.add_argument('-headless')
options.page_load_strategy = "eager"

driver = webdriver.Firefox(options=options)
driver.set_page_load_timeout(25)
driver.implicitly_wait(20)
driver.set_script_timeout(20)

# Define variables and functions
num_timeouts = 0
MAX_TIMEOUTS = 3
```

```python
def fetch_video_info(url):
    driver.get(url)

    try:
        title_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="container"]/h1/yt-formatted-string'))
        )
        title = title_element.text
    except NoSuchElementException:
        print("Error fetching video title.")
        title = None
    except TimeoutException:
        print("Timeout occurred while fetching video title.")
        title = None

    try:
        views_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="count"]/yt-view-count-renderer/span[1]'))
        )
        views = views_element.text
    except NoSuchElementException:
        print("Error fetching video views.")
        views = None
    except TimeoutException:
        print("Timeout occurred while fetching video views.")
        views = None

    if title and views:
        return title, views
    else:
        print(f"Error fetching video information. Title: {title}, Views: {views}")
        return None, None
```

```python
url = str(url_input.value)
driver.get(url)

title_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="container"]/h1/yt-formatted-string'))
        )

print(title_element.text)
```

```python
# Example usage
url = "https://www.youtube.com/watch?v=dQw4w9WgXcQ"  # Replace with the desired YouTube video URL
title, views = fetch_video_info(url)

if title and views:
    print(f"Title: {title}\nViews: {views}")
else:
    print("Failed to fetch video information.")


# Close the driver
driver.quit()
```

```python
import re
from IPython.display import Javascript
from functools import partial
import ipywidgets as widgets

def split_text(text, num_words):
    words = re.findall(r'\b\w+\b', text)
    return [' '.join(words[i:i+num_words]) for i in range(0, len(words), num_words)]

# Function to copy text to clipboard
def copy_to_clipboard(text):
    display(Javascript(f'navigator.clipboard.writeText("{text}")'))

# Textarea for input
text_input = widgets.Textarea(
    value='',
    placeholder='Enter your multiline string here',
    description='Text:',
    disabled=False
)

# Integer slider to select the number of words per chunk
num_words_slider = widgets.IntSlider(
    value=5,
    min=1,
    max=10000,
    step=1,
    description='Words per chunk:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

# Button to process the text
process_button = widgets.Button(
    description='Split text',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to split text',
    icon='check'
)

# Output widget to display the result
output = widgets.Output()

# Function to handle chunk button click
def on_chunk_button_click(b, chunk):
    print(chunk)
    copy_to_clipboard(chunk)

# Function to handle process button click
def on_process_button_click(b):
    with output:
        output.clear_output()
        text = sample
        #text = text_input.value
        num_words = num_words_slider.value
        chunks = split_text(text, num_words)

        buttons = []
        for i, chunk in enumerate(chunks):
            button = widgets.Button(description=f"Copy chunk {i+1}")
            button.on_click(partial(on_chunk_button_click, chunk=chunk+" "))
            buttons.append(button)

        button_grid = widgets.GridBox(buttons, layout=widgets.Layout(grid_template_columns="repeat(3, 150px)"))
        display(button_grid)

process_button.on_click(on_process_button_click)

# Display the widgets
display(text_input)
display(num_words_slider)
display(process_button)
display(output)
# 1085 wirds giid
```
